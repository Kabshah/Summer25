# BOT SORT

```python
from ultralytics import YOLO

# Load model
model = YOLO("yolo11m.pt")

# Run tracking using BoT-SORT
results = model.track(
    source="/content/datset.mp4",
    persist=True,
    save=True,
    show=True,
    tracker="botsort.yaml"   # 👈 BoT-SORT tracker
)
```

# Byte Track

```python
from ultralytics import YOLO
model = YOLO("yolo11m.pt")
results = model.track(
    source="/content/datset.mp4",
    persist=True,
    save=True,
    tracker="bytetrack.yaml"
)
```
